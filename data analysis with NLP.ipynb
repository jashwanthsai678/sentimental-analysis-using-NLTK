```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
with open('StopWords_GenericLong.txt', 'r') as f:
    stopword_list = [word.strip() for word in f.readlines()]

# Assuming you have a dataset stored in a pandas DataFrame named 'df' with a column 'Text'
# Remove stopwords from the 'Text' column
data['URL'] = data['URL'].apply(lambda x: ' '.join([word for word in x.split() if word.lower() not in stopword_list]))

# Save the modified dataset to a new file
data.to_csv('modified_wordsinlist1.csv', index=False)
```

```python
sid = SentimentIntensityAnalyzer()

# Perform text analysis and compute variables
def compute_sentiment_scores(URL):
    sentiment_scores = sid.polarity_scores(URL)
    positive_score = sentiment_scores['pos']
    negative_score = sentiment_scores['neg']
    return positive_score, negative_score

# Assuming you have a dataset stored in a pandas DataFrame named 'df' with a column 'Text'
data['Positive Score'], data['Negative Score'] = zip(*data['URL'].apply(compute_sentiment_scores))
```

```python
from textblob import TextBlob

def compute_sentiment_scores(URL):
    blob = TextBlob(URL)
    polarity_score = blob.sentiment.polarity
    subjectivity_score = blob.sentiment.subjectivity
    return polarity_score, subjectivity_score

# Assuming you have a dataset stored in a pandas DataFrame named 'df' with a column 'Text'
data['Polarity Score'], data['Subjectivity Score'] = zip(*data['URL'].apply(compute_sentiment_scores))
```
